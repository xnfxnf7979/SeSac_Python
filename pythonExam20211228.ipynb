```python
# 딕셔너리 만들기

'''
딕셔너리명 = {키:값, ...}

딕셔너리명 = dict(키1=값1,...)
딕셔너리명 = dict(zip(키 리스트, 값 리스트))
딕셔너리명 = dict([(키1,값1),(키2,값2),..])
딕셔너리명 = dict({키1:값1, 키2:값2,...})
'''

menu = {'김밥':2000,
        '떡볶이':2500,
        '어묵':2000,
        '튀김':3000}
menu

menu1 = dict(김밥=2000,
            떡볶이=2500,
            어묵=2000,
            튀김=3000)
menu1

key_list = ['김밥','떡볶이','어묵','튀김']
val_list = [2000,2500,2000,3000]
menu2= dict(zip(key_list,val_list))
menu2

menu3 = dict([('김밥',2000),('떡볶이',2500),('어묵',2000),('튀김',3000)])
menu3

menu4 = dict({'김밥':2000,
        '떡볶이':2500,
        '어묵':2000,
        '튀김':3000})
menu4
```

```python
# 함수에서 빠져나오기

# 리턴 끝나면 함수 빠져나옴
# 반환할 값이 있다면 값 반환 후 빠져나오고 없으면 걍 빠져나옴

# 정수 입력받아 0,작수,홀수 여부 리턴
def is_odd_even(n):
    if n ==0:
        res = 0
    elif n%2 ==0:
        res= '짝수'
    else:
        res = '홀수'
    return res
    '''
    if n ==0:
        return 0
    elif n%2 ==0:
        return '짝수'
    return '홀수'
    '''

is_odd_even(5)
```

```python
# 매개변수로 전달받은 수가 소수인지 아닌지 판별

def is_prime(n):
    if n <=1:
        return False
    
    for i in range(2,n):
        if n%i==0:
            return False
    return True

is_prime(2)
```

### 람다함수와 map 함수

```python
# 람다함수

# 매개변수, 수식으로 이루어진 함수
# 익명함수..이름 없음. 
# (lambda 매개변수1, 2 : 수식)(인수1,2)

def get_plus(n1,n2):
    return n1+n2
get_plus(1,2)
# 람다 로 표현
(lambda n1,n2:n1+n2)(1,2)
```

```python
# map 함수에 람다식 사용

# 리스트에 1을 더한 l2 리스트 생성
l1 = [1,2,3,4,5]
#(lambda x:x+1)

l2=list(map((lambda x:x+1),l1))
print(l2)
```

## 클래스와 객체

```python
'''
객체명.속성명 = '값'
객체명.메소드()

class 클래스명:
def __init__(self,매1,매2,..): >> 생성자
self : 메소드를 호출한 객체를 받음
'''

# 자동차 클래스와 객체

# 자동차 클래스 생성
class Car:
    def __init__(self,b,m,c):
        self.brand = b
        self.model = m
        self.color = c
        
        print(f'{self.brand}-{self.model}-{self.color}출고')
    
    def turn_on(self):
        print(f'{self.brand}에 시동 겁니다')
    def turn_off(self):
        print(f'{self.brand}에 시동 끕니다')
    def drive(self):
        print(f'{self.brand} 주행 중')
        
# 객체 생성
car1 = Car('현대','소나타','화이트')
car2 = Car('르노삼성','SM7','블랙')

# 메소드 호출
car1.turn_on()
car1.drive()
car1.turn_off()

car2.turn_on()
car2.drive()
```

```python
# df['math']의 모든 점수에 5점 더하기 + 100점 넘을 수 없음!

def p5(x):
    scroe = x+5
    if scroe >=100:
        scroe = 100
    return scroe

df_copy['math'] = df['math'].apply(p5)
df_copy
```

```python
# df['eng']의 모든 점수에 n점 더하기
def plus(x,n):
    score = x+n
    if score >=100:
        score = 100
    return score

# 1점 더하기
df_copy['eng'] = df['eng'].apply(plus,n=1)
df_copy
```

```python
# datetime 자료형 이용

import pandas as pd
df = pd.read_csv('data/birth_die.csv')

# 컬럼을 datetime 자료형으로 변경하기

# 출생, 사망 컬럼을 datetime 자료형으로 변경하기
df['출생'] = pd.to_datetime(df['출생'])
df['사망'] = pd.to_datetime(df['사망'])


# 연, 월, 일, 분기 추출하기

# 출생 컬럼 0번 인덱스의 연도
df['출생'][0].year

# 출생 컬럼의 연도
df['출생'].dt.year

df['출생'].dt.month

df['출생'].dt.day

# 분기 컬럼 만들기
df['분기']=df['출생'].dt.quarter


# 날짜 계산하기

# 생존일수 컬럼 만들기
df['생존일수'] = df['사망']-df['출생']

# 생존기간 컬럼 만들기
df['생존기간'] = df['사망'].dt.year-df['출생'].dt.year


# 요일, 월이름 추출하기
'''
컬럼.dt.strftime('%a') : 요약된 요일이름
컬럼.dt.strftime('%A') : 긴 요일이름
컬럼.dt.strftime('%w') : 숫자요일(0:일요일)
컬럼.dt.strgtime('%b') : 요약된 월이름
컬럼.dt.strftime('%B') : 긴 월이름
'''

df['출생'].dt.strftime('%b')

# 출생요일 컬럼 
df['출생요일'] = df['출생'].dt.strftime('%a')

# 출생월 컬럼 추가하기
df['출생월'] = df['출생'].dt.strftime('%b')


# datetime 자료형을 인덱스로 만들어 사용하기

# 출생 컬럼을 인덱스로 만들기
df.index = df['출생']

# 1955년 출생한 데이터 추출하기
df.loc['1955']

# 1955년 2월 출생한 데이터 추출하기
df.loc['1955-02']
```

### 그룹별 데이터 집계

```python
# groupby 메소드 이용

df = pd.read_csv('data/titanic.csv')
df = df[['Survived','Pclass','Sex','Age','Embarked']]
df = df.dropna()
df.head()
```

```python
# 그룹의 통계값 계산
'''
df.groupby(그룹기준컬럼).통계적용컬럼.통계함수
count() : 누락값을 제외한 데이터 수
size() : 누락값을 포함한 데이터 수
mean() : 평균
sum() : 합계
std() : 표준편차
min() : 최소값
max() : 최대값
sum() : 전체 합
'''
```

```python
# 객실등급별 생존 통계

# 객실등급(Pclass)별 승선자 수를 구한 결과를 데이터프레임 df1로 만들기
df1 = df.groupby('Pclass').Survived.count().to_frame()
df1
```

```python
# 객실등급(Pclass)별 생존자 수를 구한 결과를 데이터프레임 df2로 만들기
df2 = df.groupby('Pclass').Survived.sum().to_frame()
df2
```

```python
# 객실등급(Pclass)별 생존율 구한 결과를 데이터프레임 df3으로 만들기
df3 = df.groupby('Pclass').Survived.mean().to_frame()
df3
```

```python
# 객실등급(Pclass)별 탑승자수, 생존자수, 생존율 데이터프레임을 df4로 만들기
df4 = pd.concat([df1,df2,df3],axis=1)
df4.columns=['승선자수','생존자수','생존율']
df4
```

```python
# 성별 생존 통계

# 성별 승선자 수 데이터프레임을 df5로 만들기
df5 = df.groupby('Sex').Survived.count().to_frame()
df5
```

```python
# 성별 생존자 수 데이터프레임을 df6으로 만들기
df6 = df.groupby('Sex').Survived.sum().to_frame()
df6
```

```python
# 성별 생존율 데이터프레임을 df7로 만들기
df7 = df.groupby('Sex').Survived.mean().to_frame()
df7
```

### 지하철 시간대별, 역별 이용현황 분석

```python
# 시간대병 승차 현황

# 승차정보 집계 데이터 만들기
df_in_202108
```

```python
# df_in_202108 copy 해 사용
df_in_08_agg = df_in_202108.copy()
df_in_08_agg
```

```python
# 지하철역을 인덱스로 변경
df_in_08_agg.index = df_in_08_agg['지하철역']
df_in_08_agg
```

```python
# 컬럼 삭제
df_in_08_agg.drop(columns=['사용월','호선명','지하철역'],inplace=True)
df_in_08_agg
```

```python
# 행,열 합계 추가
df_in_08_agg.loc['sum'] = df_in_08_agg.apply('sum',axis=0)
df_in_08_agg
```

```python
df_in_08_agg['sum'] = df_in_08_agg.apply('sum',axis=1)
df_in_08_agg
```

```python
# 시간대별 승차건수
s_in = df_in_08_agg.loc['sum'][:-1].sort_values()
```

```python
s_in = s_in.sort_index()
s_in
```

```python
# 시각화
x=s_in.index
y=s_in.values

plt.bar(x,y)
plt.xticks(rotation=90)
plt.show()
```

```python
# 지하철역별 승차건수

# 하루동안 승차건수 많은 역
df_in_08_agg['sum'][:-1].sort_values(ascending=False).to_frame()
```
